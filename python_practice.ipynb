```python
# count the frequency of lis elements
li = [1,2,2,3,1,1]
result = {}

for i in li:
    if i not in result.keys():
        result[i] = 0
        for j in li:
            if i == j:
                result[i] += 1
print(result)
```

```python
# function to count the vowels in any word
def vowel_count(st):
    vowel_list = {"a","e","i","o","u"}
    count = 0

    for char in st:
        if char in vowel_list:
            count +=1
    return count

result = vowel_count("programming")
print(result)
```

```python
# function to check if the word is palindorm
def check_pali(word):
    if not isinstance(word, str):
        raise TypeError("enter the word")
    
    word = word.lower().strip()

    left = 0
    right = len(word) - 1

    while left < right:
        if word[left] != word[right]:
            return False
        left += 1
        right -= 1
    
    return True


result = check_pali("121")
print(result)
```

```python
# print first non repeating element from string
def non_repeating(word):
    if not isinstance(word, str):
        raise ValueError("Must enter the word")
    
    word = word.lower().strip()

    word_count = {}

    for char in word:
        word_count[char] = word_count.get(char, 0) + 1
    
    for char in word:
        if word_count[char] == 1:
            return char
    return None
result = non_repeating("aabbcdd")
print(result)
```
